## scipy

과학계산용 함수를 모아놓은 파이썬 패키지이다.   
고성능 선형대수, 함수 최적화, 신호처리, 특수한 수학함수 및 통계분포 등의 기능을 제공한다.

```python
from scipy import sparse
```

#### CSR을 이용한 희소행렬(희박행렬) 만들기

```python
# 대학선 원소는 1이고 나머지는 0인 2차원 numpy 배열을 만듦
eye = np.eye(4) 
eye
```

**CSR** : Compressed Sparse Row의 약자로 행의 인덱스를 압축하여 저장   
**COO** : Coordinate의 약자로 데이터가 놓일 행렬의 위치를 별도의 매개변수로 전달

```python
# numpy 배열을 CSR 포맷의 Scipy 희박 행렬로 변환.
# 0 이 아닌 원소만 저장된다

sparse_matrix = sparse.csr_matrix(eye)
print(sparse_matrix)
```

### COO를 이용한 희소행렬 만들기

```python
data = np.ones(4)
row_indices = np.arange(4)
col_indices = np.arange(4)
eye_coo = sparse.coo_matrix((data, (row_indices, col_indices)))

print(eye_coo)
```

# 붓꽃의 품종 분류

목표 : 어떤 품종인지 구분해놓은 측정 데이터를 이용해 새로 채집한 붓꽃의 품종을 예측하는 머신러닝 모델 개발

붓꽃 품종을 정확하게 분류한 데이터를 갖고 있으므로 **지도학습**에 속함

class : 출력될 수 있는 값(붓꽃의 종류)   
label : 특정 데이터 포인트에 대한 출력(붓꽃하나에 대한 기대 출력 = 꽃의 품종)

```python
from sklearn.datasets import load_iris
iris_dataset = load_iris()
```

```python
iris_dataset
```

```python
print("iris_dataset의 key : ",iris_dataset.keys())
```

### DESCR

데이터셋에 대한 간략한 설명을 보여준다.

```python
print(iris_dataset['DESCR'][:193] + "\n...")
```

### target_names

우리가 예측하려는 붓꽃 품종의 이름을 문자열 배열로 저장

```python
print("타깃의 이름 :",iris_dataset['target_names'])
```

### feature_names

각 특성을 설명하는 문자열 리스트

```python
print("특성의 이름 : ",iris_dataset['feature_names'])
```

### data

꽃잎의 길이와 폭, 꽃받침의 길이와 폭을 수치 값으로 가지고 있는 numpy array.

행 : 개개의 꽃   
열 : 각 꽃에서 구한 네 개의 측정치

```python
print("data type : ",type(iris_dataset['data']))

print("data의 size : ",iris_dataset['data'].shape)
```

```python
iris_dataset['data'][:5]
```

### target

붓꽃의 품종을 담은 numpy array

```python
print("target type : ",type(iris_dataset['target']))

print("target size : ", iris_dataset['target'].shape)
```

```python
print("target : \n",iris_dataset['target'])
```

```python
iris_dataset['target_names']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(
iris_dataset['data'],iris_dataset['target'],random_state=0)
```

```python
print(X_train.shape)
print(X_test.shape)
```

```python
print(y_train.shape)
print(y_test.shape)
```

## scatter plot

```python
iris_dataframe = pd.DataFrame(X_train, columns=iris_dataset.feature_names)

pd.plotting.scatter_matrix(iris_dataframe, c=y_train, figsize=(15,15), marker="o",hist_kwds={'bins':20}, s=60, alpha=.8, cmap=mglearn.cm3)
```

## k-최근접 이웃 알고리즘

K-Nearest Neighbors.    
새로운 데이터 포인트에 대한 예측이 필요하면 알고리즘은 새 데이터 포인트에서 가장 가까운 train data 포인트를 찾는다.   
k라면, 가장 가까운 'k개'의 이웃을 찾고 이 이웃들의 클래스 중 빈도가 가장 높은 클래스를 예측값으로 사용한다.   

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=1)
```

```python
knn.fit(X_train, y_train)
```

## Predict

```python
X_new = np.array([[5,2.9,1,0.2]])
print(X_new.shape)
```

```python
prediction = knn.predict(X_new)
print("예측 : ",prediction)
print("예측한 타겟 이름 : ",iris_dataset['target_names'][prediction])
```

## 모델 평가하기

```python
y_pred = knn.predict(X_test)
print("test set에 대한 예측값 : \n",y_pred)
```

```python
print("정확도 : {:.2f}".format(np.mean(y_pred == y_test)))
```

```python
print("정확도 : {:.2f}".format(knn.score(X_test, y_test)))
```
